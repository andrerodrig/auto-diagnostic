```python
import cv2
from skimage.filters import roberts, sobel, scharr, prewitt
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow


#Processamento da imagem 
img = cv2.imread('/content/drive/MyDrive/leaves/063.jpg')
green_image = img.copy()
green_image[:, :, 0] = 0
green_image[:, :, 2] = 0
img_g = img[:,:,1]
#G_img = cv2_imshow(img_g) 
```

```python
#aplicando o sobel 
op_sobel = sobel(img_g)
```

```python
#Plotagem 
fig, axes = plt.subplots(nrows=2, ncols=2, sharex=True, sharey=True, figsize=(12, 12))

ax = axes.ravel()

ax[1].imshow(op_sobel, cmap=plt.cm.gray)
ax[1].set_title('Sobel')

for a in ax:
    a.axis('off')

plt.tight_layout()
plt.show()
```
